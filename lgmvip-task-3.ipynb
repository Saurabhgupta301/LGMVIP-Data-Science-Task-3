# Generate Pencil Sketch from Photo in Python

## <font color = blue> Import Libraries </font>

**` OpenCV is the only library which is needed for the project.
We will also be using matplotlib library for some visualizations which is discussed later.`**

```python
import cv2
import matplotlib.pyplot as plt
```

## <font color = blue> Read Photo </font>

**`The following command can be used to read image using OpenCV.`**

```python
img=cv2.imread("Neeraj-chopra.jpg")
```

**`This command reads the file Neeraj-chopra.jpg located in the current folder and stores in memory as img.`**

## <font color = blue> Show Image using OpenCV </font>

**`Displaying image using OpenCV is not very straight forward. The following command can be used to display the photo.`**

## <font color = blue> Using cvtColor method of OpenCV. </font>

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

## Convert Image to a Pencil Sketch

**`Converting a photo to pencil sketch involves the following steps:`**

## <font color = blue> Step 1: Convert to Grey Image </font>

**`Using cvtColor function of OpenCV.`**

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(grey_img)
plt.axis(False)
plt.show()
```

## <font color = blue> Step 2: Invert ImageStep 2: Invert Image </font>

**`Inverting the image can be done by either of the following method.`**

```python
invert_img=cv2.bitwise_not(grey_img)
#invert_img=255-grey_img
plt.imshow(invert_img)
plt.axis(False)
plt.show()
```

## <font color = blue> Step 3: Blur image </font>

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
plt.imshow(blur_img)
plt.axis(False)
plt.show()
```

## <font color = blue> Step 4: Invert Blurred Image </font>

**`Repeat step 2`**

```python
invblur_img=cv2.bitwise_not(blur_img)
#invblur_img=255-blur_img
plt.imshow(invblur_img)
plt.axis(False)
plt.show()
```

## <font color = blue> Step 5: Sketch </font>

**`The sketch can be obtained by performing bit-wise division between the grayscale image and the inverted-blurred image.`**

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
plt.imshow(sketch_img)
plt.axis(False)
plt.show()
```

## <font color = blue>Step 6: Save Sketch </font>



```python
cv2.imwrite('sketch.png', sketch_img)
```

## <font color = blue> Step 7: Display sketch </font>

```python
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```

# Original Image vs Sketch

**`We can display the original image and sketch side by side for comparison.`**


```python
# ORIGINAL IMAGE
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
plt.imshow(RGB_img)
plt.axis('off')

# SKETCH IMAGE
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
